# array operations

```python
a1= np.arange(12).reshape(3,4)
a2= np.arange(12,24).reshape(3,4)

a1
```

```python
#scalar operations
# arithmetic

print(a1 * 2)

print(a2 + 2)
```

```python
#relational
a2>15
```

# Array functions

```python
a1=np.random.random((3,3))
a1=np.round(a1*100)
a1
```

```python
#max/min/sum/prod
print(np.max(a1))
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))
```

```python
# 0-> col, 1-> row
#max/min/sum/prod in particular row/col
print(np.max(a1,axis=1))
print(np.max(a1,axis=0))
```

```python
#mean/median/std/var
print(np.mean(a1))
print(np.median(a1))
print(np.std(a1))
print(np.var(a1))   #for particulars rows & col too
```

```python
# dot product
a2= np.arange(12).reshape(3,4)
a3= np.arange(12,24).reshape(4,3)

np.dot(a2,a3)
```

```python
#lof & exponenets
print(np.exp(a1))
```

```python
#round/floor/ceil

print(np.round(np.random.random((2,3))*100))
print(np.floor(np.random.random((2,3))*100))
print(np.ceil(np.random.random((2,3))*100))
```
