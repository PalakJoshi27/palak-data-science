```python
#Reverse a string
s = input("Enter a string: ")
print(s[::-1])
```

```python
#Check whether a string is a palindrome.
s=input("enter a string=")
if s ==s[::-1]:
  print("string is palindrome")
else:
  print("string is not palindrome")

```

```python
#Count the number of vowels and consonants in a string.

s = input("enter a string = ")
vowels = "aeiouAEIOU"
v = 0
c = 0

for i in s:
    if i.isalpha():
        if i in vowels:
            v += 1
        else:
            c += 1

print("Vowels:", v)
print("Consonants:", c)
```

```python
'''Convert a string to:

uppercase

lowercase

title case'''

s = input("enter a string = ")
print(s.upper())
print(s.lower())
print(s.title())
```
