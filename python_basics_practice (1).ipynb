## Conditional statements

```python
## 1. check whether a no. is odd or even
n=int(input("enter no. to check="))

if n%2==0:
  print("the no. is even")
else:
  print("the no. is odd")
```

```python
## 2. check whether a no. is positive ,negative or zero
n= int(input("enter no. to check="))
if (n>0):
   print("the no. is positive")
elif(n<0):
   print("the no. is negative")
else:
   print("the no. is zero")
```

```python
## 3. find largest of three no.s
a= int(input("enter no.="))
b= int(input("enter no.="))

c= int(input("enter no.="))
if (a>b and a>c):
   print("a is largest")
elif(b>a and b>c):
   print("b is largest")
else:
   print("c is largest")
```
