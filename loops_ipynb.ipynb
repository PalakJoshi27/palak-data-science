## **Operators in Python**

 1.Arithmetic Operator

 2.Relational Operators

3.Logical Operators

4.Bitwise Operators

5.Assignment Operators

6.Membership Operators




```python
# Arithmetric Operators
print(5+6)

print(5-6)

print(5*6)

print(5/2)

print(5//2)

print(5%2)

print(5**2)
```

```python
# Relational Operators
print(4>5)

print(4<5)

print(4>=4)

print(4<=4)

print(4==4)

print(4!=4)
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

print(~3)

print(4 >> 2)

print(5 << 2)
```

```python
# Assignment Operators

a=2
a+=2 # a=a+2
print(a)
```

```python
# Membership Operators

# in/not in

print('D' not in 'Delhi')

print(1 in [2,3,4,5,6])
```

# Loops in python

```python
# print the multiplication table of a no. entered by user
a= int(input("Enter a no."))
i=1
while (i<11):
  print(i*a)
  i +=1
```

```python
# while with else

i=1
while (i<5):
  print(i)
  i+=1
else:
  print("limit cross")
```

```python
# for loop

a=[1,2,3,4,5]
for i in a:
  print(i)
```

# Nested loops

```python
for i in range(1,5):
  for j in range(1,4):
    print(i,j)
```

```python
'''print this
*
**
***
**** '''

for i in range(1,5):
  for j in range(1,i+1):
    print('*',end='')
  print()

```

# Loops control statements

*   break
*   continue
*   pass


```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
# Continue
for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

```python
for i in range(1,10):
  pass
   # it will run w/o raising any error
```
